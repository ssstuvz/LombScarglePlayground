Playing with Fourier spectrum of not evenly distributed signal
===============================================================

Introduction
------------

It can happen that you will end up with a signal that was measured in not equally spaced moments of time. Say time can be (in seconds or whatever): 1,2, 10, 3, 5, 3.3, 1.7, even completely random. For me, this problem arose when trying to calculate the radiation of particles moving inside plasma.

How do we find the Fourier spectrum of such a signal? Apparently, there was a lot of research done in this area, and I have found that mainly the method refered to as [Lomb-Scargle](http://en.wikipedia.org/wiki/Least-squares_spectral_analysis) is used. Here are some examples of using it in Python and comparison with FFT and interpolation method. 

Creating model signal
---------------------
Let's create some signal that we will analyze. I use `ipython notebook --pylab inline` for this post so that `numpy, matplotlib` are already imported. Python function `signal_function(t)` describes the signal: it is 3 different Gaussian pulses with 3 different frequencies: 1, 3 and 5. 
Function `create_test_signal_random` generates random values for `t` in the range from `t_start` to `t_end` and returns both the time axis and the generated signal.
Function `create_test_signal_even` created evenly spaced version of the signal.


```python
def signal_function(t):
    y=exp(-(t-50*2*pi)**2/2.0/50**2)*sin(t)+exp(-(t-45*2*pi)**2/2.0/50**2)*sin(3*t)+\
    1.*exp(-(t-20*2*pi)**2/2.0/15**2)*sin(5*t)
    return y

def create_test_signal_random(t_start, t_end, NofTS=10000):
    seed=random.randint(10000)
    random.seed(seed)
    t=t_start + random.rand(NofTS)*(t_end-t_start)
    y=signal_function(t)  
    return (t,y)

def create_test_signal_even(t_start, t_end, NofTS=10000):

    t=linspace(t_start, t_end, NofTS)
    y=signal_function(t)
    
    return (t,y)

t_start=0.0*2*pi
t_end=90*2*pi

NofTS = 10000

t,y=create_test_signal_random(t_start, t_end, NofTS)
tlin,ylin=create_test_signal_even(t_start, t_end, NofTS)

plot(tlin/2/pi, ylin)
xlabel('Time, arb. units', fontsize=16)
ylabel('Signal, arb. units', fontsize=16)
```

Plot above shows the signal. Now let's find the Fourier spectrum of it.

Fourier spectrum of the signal
------------------------------

Scipy already has Lomb-Scargle method implemented, so it is a one-liner to generate the spectrum.
`f` is the vector of frequencies where the spectrum will be calculated (in the example below from 0.0001 to 6)

Below 5 different versions of spectrum calculation are performed:
<br>
`pgram_random` is the power spectrum of randomly distributed signal using Lomb-Scargle.<br>
`pgram_lin` is the Lomb-Scargle of evenly distributed signal.<br>
`spec_fft` is the FFT (numpy) of the evenly distributed signal.<br>
`spec_random` is simple summation calculation of the randomly-distributed spectrum. <br>
`spec_lin` is simple summation calculation of the evenly distributed spectrum.

```python
import scipy.signal as signal
f=linspace(0.0001, 6, 1000)

# Lomb-Scargle periodogram of random and evenly spaced signals
pgram_random = signal.lombscargle(t, y, f)
pgram_lin = signal.lombscargle(tlin, ylin, f)

# FFT of evenly spaced signal
spec_fft=fft.fft(ylin)
w=fft.fftfreq(NofTS, d=(tlin[1]-tlin[0])/2/pi)

# simple summation calculation 
spec_random=zeros(1000)
spec_lin=zeros(1000)

n=0
for omega in f:
    speccos=np.sum(y*cos(omega*t))
    specsin=np.sum(y*sin(omega*t))
    spec_random[n]=(speccos**2+specsin**2)
    speccos=np.sum(ylin*cos(omega*tlin))
    specsin=np.sum(ylin*sin(omega*tlin))
    spec_lin[n]=(speccos**2+specsin**2)
    n=n+1
```

Now let's plot them together. 

```python
semilogy(f,pgram_random/max(pgram_random), 'k', lw=2, label='Lomb-Scargle random')
semilogy(f,pgram_lin/max(pgram_lin), 'b--', lw=2, label='Lomb-Scargle even')
plot(w, abs(spec_fft**2)/max(abs(spec_fft**2)), 'r', lw=2, label='FFT even')

xlim(0,6)
ylim(1e-6,1)

hlines(1.0/(NofTS), 0, 10)
hlines(1/sqrt(NofTS), 0, 10)
    
plot(f, spec_random/max(spec_random), 'g-.', lw=2, label='Simple sum random')
plot(f, spec_lin/max(spec_lin), 'c--', lw=2, label='Simple sum even')
xlabel('Frequency, arb. units', fontsize=16)
ylabel('Normalized power spectrum', fontsize=16)
legend(bbox_to_anchor=(1.6, 1.0))
```

Basically, all methods give similar results. There is a huge difference between the evenly and randomly distributed signals though. One can see that there is additional noise in the spectrum introduced by uneven random spacing and the level of this noise is between `1./NofTS` and `1./sqrt(NofTS)`. One can play around with `NofTS` and see how the noise level changes. 

Interpolation
-------------

Another idea would be to interpolate the randomly distributed signal to even grid and do the FFT. Here, we interpolate on the grid that has same number of time steps:

```python
#interpolation
data=vstack([t,y])
data=data.T
data=data[np.argsort(data[:,0])]

t=data[:,0]
y=data[:,1]

plot(t/2/pi,y)

NofTS_interp = NofTS

t_interp = linspace(t_start, t_end, NofTS)
y_interp = np.interp(t_interp, t, y)

plot(t_interp/2/pi, y_interp)

sp_interp=fft.fft(y_interp)
f_interp=fft.fftfreq(NofTS_interp, d=(t_interp[1]-t_interp[0])/2/pi)
```

Let's compare the spectra for interpolation FFT and Lomb-Scargle:

```python
semilogy(f,pgram_random/max(pgram_random))
plot(f_interp, abs(sp_interp)**2/max(abs(sp_interp)**2))


xlim(0,6)
ylim(1e-6,1)

hlines(1.0/(NofTS), 0, 10)
hlines(1/sqrt(NofTS), 0, 10)
```

One can see that in the case of the interpolation the noise level goes down. I am still not sure which method is the best, this is first time I do it, so I am really looking for your comments/questions/suggestions :)
